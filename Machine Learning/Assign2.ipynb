Q1) Load the dataset into a pandas dataframe and display the first 5 lines of the dataset along with the column headings. Note that the text file does not have the headers, which means you will have to add them to the dataframe.

```python
import pandas as pd
dataset = pd.read_csv('data.txt',sep='\s+', 
                      names=["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX","PTRATIO","B", "LSTAT", "MEDV"])
dataset.head()
```

Q2) Split the dataset into training (70%) and testing set (30%).

```python
import numpy as np
from sklearn.model_selection import train_test_split
#extract x (all columns minus target variable) & y (target variable)
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,13].values

#split into test/train
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

#print(x_train.shape)
#print(x_test.shape)
#print(y_train.shape)
#print(y_test.shape)
```

Q3) Model 1: Using scikit learn, build a Linear Regression model with all the variables.

```python
#Fit Linear Regression to training set
from sklearn.linear_model import LinearRegression
lrObj =LinearRegression()
lrObj.fit(x_train, y_train)
```

Q4) What are the weight parameters (including the intercept) you get for Model 1.

```python
print("Intercept: " , lrObj.intercept_)
print("Weight parameters: ", lrObj.coef_)
```

Q5) Use Model 1 to make a prediction on the test set. Calculate mean squared error.

```python
y_pred=lrObj.predict(x_test)

# mean squared error
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)
print('MSE = ', MSE)
print(y_pred)
```

Q6) Model 2: Build a linear regression model with all the variables using Normal Equations method.

```python
#extract x (all columns minus target variable) & y (target variable)
x1 = dataset.iloc[:,:-1].values
y2 = dataset.iloc[:,13:].values
#add column of ones to x1 for intercept
x2 = np.append(arr=np.ones((506,1)).astype(int), values=x1, axis=1)

#split into test/train
x2_train,x2_test,y2_train,y2_test = train_test_split(x2,y2,test_size=0.3)

#Use normal equation method
#w = (xtx)i * xty
transpose=x2_train.T
mult1=np.dot(transpose, x2_train)
inverse=np.linalg.inv(mult1)
mult2=np.dot(inverse,transpose)
normal_eq = np.dot(mult2, y2_train)
print(normal_eq.shape)
```

Q7) What are the weight parameters you get for Model 2.

```python
print(normal_eq)
```

Q8) Use Model 2 to make a prediction on the test set. Calculate mean squared error.

```python
#prediction
y_pred2= x2_test.dot(normal_eq)

# mean squared error
MSE2 = mean_squared_error(y2_test, y_pred2)
print('MSE = ', MSE2)
print(y_pred2)
```
